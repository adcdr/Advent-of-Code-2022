# Part A

```python
sections = []

with open('data/day-4.data') as f:
    lines = f.readlines()

    for line in lines:
        pair_sections = line.strip().split(',')

        section_1_range = pair_sections[0].split('-')
        section_2_range = pair_sections[1].split('-')

        section_1_range = list(map(int, section_1_range))
        section_2_range = list(map(int, section_2_range))

        sections.append([section_1_range, section_2_range])
        print(pair_sections)
```

```python
len(sections)
```

```python
fully_contained_count = 0

for section in sections:
    if section[0][0] >= section[1][0] and section[0][1] <= section[1][1]:
        fully_contained_count += 1

    elif section[1][0] >= section[0][0] and section[1][1] <= section[0][1]:
        fully_contained_count += 1
```

```python
fully_contained_count
```

# Part 2

```python
partially_contained_count = 0

for section in sections:
    if section[1][0] <= section[0][0] <= section[1][1]:
        partially_contained_count += 1

    elif section[1][0] <= section[0][1] <= section[1][1]:
        partially_contained_count += 1

    elif section[0][0] <= section[1][0] <= section[0][1]:
        partially_contained_count += 1

    elif section[0][0] <= section[1][1] <= section[0][1]:
        partially_contained_count += 1
```
