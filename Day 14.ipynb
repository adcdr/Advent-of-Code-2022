# Part 1

```python
paths = []
min_x = min_y = max_x = max_y = None

with open('data/day-14.data') as f:
    for line in f.readlines():
        line_parts = line.strip().split(' -> ')
        coordinates = []

        for coordinate in line_parts:
            coordinate_parts = coordinate.split(',')
            x = int(coordinate_parts[0])
            y = int(coordinate_parts[1])

            if min_x is None: min_x = x
            elif x < min_x: min_x = x

            if max_x is None: max_x = x
            elif x > max_x: max_x = x

            if min_y is None: min_y = y
            elif y < min_y: min_y = y

            if max_y is None: max_y = y
            elif y > max_y: max_y = y

            coordinates.append([int(coordinate_parts[0]), int(coordinate_parts[1])])

        paths.append(coordinates)

print('min_x', min_x)
print('max_x', max_x)
print('min_y', min_y)
print('max_y', max_y)
```

```python
AIR_SYMBOL = '.'
WALL_SYMBOL = '#'
SAND_SYMBOL = 'o'

X_OFFSET = min_x - 209



def generate_grid(floor_exists=None):
    grid = [[AIR_SYMBOL] * (max_x - min_x + 521) for i in range(max_y + 4)]

    for path in paths:
        for i in range(1, len(path)):
            x_start = path[i-1][0] + 1
            y_start = path[i-1][1]

            x_delta = path[i][0] - x_start + 1
            y_delta = path[i][1] - y_start

            x_range = y_range = range(0)

            if x_delta > 0: x_range = range(x_delta + 1)
            if x_delta < 0: x_range = range(x_delta, 1)
            if y_delta > 0: y_range = range(y_delta)
            if y_delta < 0: y_range = range(y_delta, 1)

            for x in x_range:
                grid[y_start][x_start + x - X_OFFSET] = WALL_SYMBOL

            for y in y_range:
                grid[y_start + y][x_start - X_OFFSET] = WALL_SYMBOL

    if floor_exists:
        for i in range(len(grid[0])):
            grid[-2][i] = WALL_SYMBOL

    return grid



grid = generate_grid()

for row in grid:
    print(row)
    print
```

```python
import time


SAND_SOURCE = [500 - X_OFFSET + 1, 0]

is_falling = False
sand_coords = [SAND_SOURCE[0], SAND_SOURCE[1]]
units_of_sand = 0
grid = generate_grid()

while True:
    if not is_falling:
        sand_coords = [SAND_SOURCE[0], SAND_SOURCE[1]]
        prev_sand_coords = [SAND_SOURCE[0], SAND_SOURCE[1]]
        units_of_sand += 1
        is_falling = True
    else:
        if grid[sand_coords[1]+1][sand_coords[0]] == AIR_SYMBOL:
            prev_sand_coords = [sand_coords[0], sand_coords[1]]
            sand_coords[1] += 1
        elif grid[sand_coords[1]+1][sand_coords[0]-1] == AIR_SYMBOL:
            prev_sand_coords = [sand_coords[0], sand_coords[1]]
            sand_coords[1] += 1
            sand_coords[0] -= 1
        elif grid[sand_coords[1]+1][sand_coords[0]+1] == AIR_SYMBOL:
            prev_sand_coords = [sand_coords[0], sand_coords[1]]
            sand_coords[1] += 1
            sand_coords[0] += 1
        else:
            is_falling = False

        grid[prev_sand_coords[1]][prev_sand_coords[0]] = AIR_SYMBOL
        grid[sand_coords[1]][sand_coords[0]] = SAND_SYMBOL

    # if sand_coords[1] == 0:
    #     print('full', units_of_sand)
    #     break

    if sand_coords[1] > max_y - 1:
        print('overflowing', units_of_sand - 1)

        # for row in grid:
        #     print(row)

        break

    # time.sleep(0.05)
    #
    # for row in grid:
    #     print(row)
    #
    # print()

# grid
```
