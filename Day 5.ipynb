#  Part A

```python
stacks = [
    ['F', 'H', 'B', 'V', 'R', 'Q', 'D', 'P'],
    ['L', 'D', 'Z', 'Q', 'W', 'V'],
    ['H', 'L', 'Z', 'Q', 'G', 'R', 'P', 'C'],
    ['R', 'D', 'H', 'F', 'J', 'V', 'B'],
    ['Z', 'W', 'L', 'C'],
    ['J', 'R', 'P', 'N', 'T', 'G', 'V', 'M'],
    ['J', 'R', 'L', 'V', 'M', 'B', 'S'],
    ['D', 'P', 'J'],
    ['D', 'C', 'N', 'W', 'V']
]
```

```python
operations = []

with open('data/day-5.data') as f:
    lines = f.readlines()

    for line in lines:
        line_parts = line.strip().split(' ')
        operations.append([int(line_parts[1]), int(line_parts[3]), int(line_parts[5])])
```

```python
len(operations)
```

```python
for operation in operations:
    for i in range(operation[0]):
        stacks[operation[2] - 1].append(stacks[operation[1] - 1].pop())
```

```python
for i in range(len(stacks)):
    print(stacks[i][len(stacks[i]) - 1], end='')
```

```python
stacks = [
    ['F', 'H', 'B', 'V', 'R', 'Q', 'D', 'P'],
    ['L', 'D', 'Z', 'Q', 'W', 'V'],
    ['H', 'L', 'Z', 'Q', 'G', 'R', 'P', 'C'],
    ['R', 'D', 'H', 'F', 'J', 'V', 'B'],
    ['Z', 'W', 'L', 'C'],
    ['J', 'R', 'P', 'N', 'T', 'G', 'V', 'M'],
    ['J', 'R', 'L', 'V', 'M', 'B', 'S'],
    ['D', 'P', 'J'],
    ['D', 'C', 'N', 'W', 'V']
]
```

```python
def print_stacks():
    for i, stack in enumerate(stacks):
        print(i, stack)
```

```python
len(operations)
```

```python
print_stacks()

for j, operation in enumerate(operations):
    temp_stack = []

    print()
    print(j, operation)
    print()

    for i in range(operation[0]):
        temp_stack.append(stacks[operation[1] - 1].pop())

    for i in range(len(temp_stack)):
        stacks[operation[2] - 1].append(temp_stack.pop())

    print_stacks()
```

```python
for i in range(len(stacks)):
    print(stacks[i][len(stacks[i]) - 1], end='')
```
